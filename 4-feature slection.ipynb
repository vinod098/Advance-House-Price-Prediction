```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#for feature selection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
dataset=pd.read_csv('3-feature_data.csv')
```

```python
dataset.head()
```

```python
y_train=dataset[['SalePrice']]
x_train=dataset.drop(['Id','SalePrice'],axis=1)
```

```python
#aapply feature selection
#first specify lasso regression and slect a suitable alpha
feature_sel_model=SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(x_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
# selected feature
selected_feature=x_train.columns[(feature_sel_model.get_support())]

print('total feature :',x_train.shape[1])
print('selected feature :',len(selected_feature))
```

```python
selected_feature
```

```python
x_train=x_train[selected_feature]
x_train.head()
```

```python
x_train.to_csv('5-Final_data',index=False)
```
