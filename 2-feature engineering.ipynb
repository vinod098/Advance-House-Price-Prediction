```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_excel('Data.xlsx')
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dataset,dataset['SalePrice'],test_size=0.1,random_state=0)
```

```python
#lets capture all the nan value
#first lets handle the categorical feature which are missing

feature_nan=[feature for feature in dataset.columns  if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

for feature in feature_nan:
    print('{}:{}% missing value'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
# replacing nan value with a new label

def replace_cat_nan(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna('missing')
    return data

dataset=replace_cat_nan(dataset,feature_nan)
dataset[feature_nan].isnull().sum()
```

```python
dataset.head()
```

```python
# nan in numerical
num_feature_nan=[feature for feature in dataset.columns  if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']

for feature in num_feature_nan:
    print('{}:{}% missing value'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
# here we replace nan with median

for feature in  num_feature_nan:
    median_val=dataset[feature].median()
    dataset[feature+'_nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_val,inplace=True)
    
dataset[num_feature_nan].isnull().sum()    
```

```python
dataset.head(10)
```

```python
# temporal variables (date time variable)
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
  
```

```python
dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
dataset.head()
```

```python
#most of numerical variable are skewed in data analysis , will perform log normal distribution

num_feature=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']

for feature in num_feature:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

```python
# here is 20 to 30 categorical feature
# rare categorical feature means we will remove those category which is present less than 1%.

categorical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
categorical_feature
```

```python
for feature in categorical_feature:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where (dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(31)
```

```python
# catgorical variables
for feature in categorical_feature:
    label_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    label_ordered={k:i for i,k in enumerate(label_ordered,0)}
    dataset[feature]=dataset[feature].map(label_ordered)
```

```python
dataset.head()
```

```python
scaling_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePerice'] ]
print(len(scaling_feature))
scaling_feature
```

```python
# feature scaling
feature_scale=[feature for feature in dataset.columns if feature not in['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
scaler.transform(dataset[feature_scale])
```

```python
data=pd.concat([dataset[['Id','SalePrice']].reset_index(drop=True),pd.DataFrame(scaler.transform(dataset[feature_scale]),columns=feature_scale)],axis=1)
```
